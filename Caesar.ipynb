```python
'''
f0 - файл войны и мира
f1 - файл главы
s0 - текст войны и мира 
s1 - текст главы
alpha - алфавит
key - сдвиг
slovar0 - частоты войны и мира
slovar1 - частоты главы

news1 - новый текст главы

news2 - для пар символов 

'''

'''Инициализация файлов и создание словаря с частотам войны и мира'''

f0 = open('voina-i-mir.txt', 'r')
f1 = open('first.txt', 'r')
#f1 = open('qwe.txt', 'r')
alpha = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'

s0 = f0.read().lower()
s1 = f1.read().lower()

key = 5

slovar0 = []

for i in alpha:
    slovar0.append([i, s0.count(i)/len(s0)])
    print((i, s0.count(i)/len(s0)))
```

```python
'''Создание зашифрованного текста с ключем в переменные news1 и news2'''

news1 = list(s1)
for i in range(len(news1)):
    if news1[i] in alpha:
        news1[i] = alpha[(alpha.index(news1[i]) + key) % len(alpha)]
        #print(alpha[(alpha.index(news1[i]) + key) % len(alpha)])
print(s1.lower(), '\n\n\n')
news1 = ''.join(news1)
print(news1)
news2 = news1

```

```python
'''Частоты зашифрованного текста'''

slovar1 = []
for i in alpha:
    slovar1.append([i, news1.count(i)/len(news1)])
    print((i, news1.count(i)/len(news1)))
```

```python
'''Сопоставление частот и просмотр разницы сдвига'''

for i in range(len(slovar0)):
    print(slovar0[i] , slovar1[i])
slovar0 = sorted(slovar0,key=lambda x: x[1])
slovar1 = sorted(slovar1,key=lambda x: x[1])
print('\n\n\n' , alpha)
for i in range(len(slovar0)):
    print(slovar0[i][0] , slovar1[i][0], (alpha.index(slovar1[i][0])-alpha.index(slovar0[i][0])))
```

```python
'''сопоставление нового алфавита'''
import numpy as np
nslovar0 = np.array(slovar0).transpose().tolist()[0]
nslovar1 = np.array(slovar1).transpose().tolist()[0]
```

```python
'''расшифровка текста на основе нового алфавита'''
print(nslovar0 , nslovar1)
news1 = list(news1)
for i in range(len(news1)):
    if news1[i] in nslovar0:
        #print(news1[i])
        news1[i] = nslovar0[nslovar1.index(news1[i])]
        
print(''.join(news1))
```

```python
'''Поиск частот пар символов для полного текста и зашифрованной части'''

#print(len(alpha))
#alpha += ' '
#print(len(alpha))
newAlpha0 = []
newAlpha1 = []
for i in range(len(alpha)):
    for j in range(len(alpha)):
        newAlpha0.append(alpha[i] + alpha[j])
#print(newAlpha0)        
newAlpha1 = newAlpha0.copy()
for i in range(len(newAlpha0)):
     newAlpha0[i] = ( [ newAlpha0[i] , (s0.count(newAlpha0[i])/len(s0))] )
#print(sorted(newAlpha0,key=lambda x: x[1]))
for i in range(len(newAlpha0)):
     newAlpha1[i] = ( [ newAlpha1[i] , (news2.count(newAlpha1[i])/len(news2))] )
#print('\n\n\n\n\n\n',sorted(newAlpha1,key=lambda x: x[1]))
```

```python
'''вывод отсортированных частот пар символов'''
newAlpha0 = sorted(newAlpha0,key=lambda x: x[1])[::-1]
newAlpha1 = sorted(newAlpha1,key=lambda x: x[1])[::-1]

for i in range(50):
    print(newAlpha0[i] , newAlpha1[i])
```

```python
'''Сопоставление частот первых пар для зашифрованного текста'''
para0 = []
para1 = []
size = 5
for i in range(size):
    ite = -1
    m = 100
    for j in range(len(newAlpha0)):
        temp = abs(newAlpha1[i][1] - newAlpha0[j][1])
        if temp < m:
            ite = j
            m = temp
    para0.extend(list(newAlpha0[ite][0]))
    para1.extend(list(newAlpha1[i][0]))
print(para0, para1)
```

```python
'''сопоставляем разницу алфавитов'''
na0 = para0.copy()
na1 = para1.copy()
# for i in range(len(newAlpha0)):
#     na0.extend(list(newAlpha0[i][0]))
#     na1.extend(list(newAlpha1[i][0]))
na0 = na0[::-1]
na1 = na1[::-1]
i = 0
while(i < len(na0)):
    if na0.count(na0[i]) > 1:
        na0.pop(i)
        na1.pop(i)
        continue
    i+=1   
#print(na0, na1)
i = 0
while(i < len(na1)):
    if na1.count(na1[i]) > 1:
        na1.pop(i)
        na0.pop(i)
        continue
    i+=1  
print(na0, na1)
nslovar2 = nslovar1.copy()
#print(nslovar2)
for i in range(len(na0)):
    if na0[i] in nslovar0:
        #print(nslovar2[nslovar0.index(na0[i])], na1[i])
        nslovar2[nslovar2.index(na1[i])] = nslovar2[nslovar0.index(na0[i])]
        nslovar2[nslovar0.index(na0[i])] = na1[i]
    
#print(nslovar0, nslovar2)
for i in range(len(nslovar0)):
    print(nslovar0[i] , nslovar2[i], (alpha.index(nslovar2[i])-alpha.index(nslovar0[i])))
```
